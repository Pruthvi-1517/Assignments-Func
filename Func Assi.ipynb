What is the difference between a function and a method in Python?

A function is a block of code that performs a specific task and is defined independently.  It's called directly by its name.   
A method is also a block of code that performs a specific task, but it is associated with an object (a class instance).  It's called using the object name, a dot, and then the method name.   
Example:
Function: def my_function(x): return x + 1 (called as my_function(5))
Method: class MyClass: def my_method(self): return "Hello" (called as obj = MyClass(); obj.my_method())

Explain the concept of function arguments and parameters in Python.

Parameters are the variables listed inside the parentheses in the function definition.  They act as placeholders for the values that will be passed to the function.   
Arguments are the actual values that are passed to the function when it is called.  These values are assigned to the parameters.   
Example:
Python

def greet(name, greeting):  # name and greeting are parameters
    print(f"{greeting}, {name}!")

greet("Alice", "Hello")  # "Alice" and "Hello" are arguments

What is the purpose of the return statement in a Python function?

The return statement is used to exit a function and send a value back to the caller of the function.    
If a function doesn't have a return statement, or if the return statement doesn't have a value, the function returns None by default.    
Example:
Python

def add(x, y):
    return x + y  # Returns the sum of x and y

What are iterators in Python and how do they differ from iterables?

An iterable is an object that can be looped over, meaning you can traverse through its elements.  Examples include lists, tuples, strings, and dictionaries.   
An iterator is an object that enables you to traverse through an iterable object element by element.  Iterators have a __next__() method that returns the next element in the sequence, and they raise a StopIteration exception when there are no more elements.    
All iterators are iterables, but not all iterables are iterators. You can get an iterator from an iterable using the iter() function.

Explain the concept of generators in Python and how they are defined.

Generators are a special type of function that returns an iterator.    
They use the yield keyword to produce a sequence of values.    
When a generator function is called, it doesn't execute the function body immediately. Instead, it returns a generator object.   
Example:
Python

def my_generator(n):
    for i in range(n):
        yield i

What are the advantages of using generators over regular functions?

Memory efficiency: Generators produce values on the fly, one at a time, and don't store the entire sequence in memory.  This is very efficient for large datasets.   
Lazy evaluation: Generators compute values only when needed.    
Improved readability: Generator code can be cleaner and easier to understand for certain tasks.

What is a lambda function in Python and when is it typically used?

A lambda function is a small, anonymous function defined using the lambda keyword.    
It can have any number of arguments but can only have one expression.    
It is typically used for short, simple operations where defining a full function is unnecessary.    
Example: square = lambda x: x**2

Explain the purpose and usage of the map() function in Python.

The map() function applies a given function to each item of an iterable (like a list) and returns an iterator of the results.    
It takes two arguments: the function to apply and the iterable.    
Example:
Python

numbers = [1, 2, 3]
squared_numbers = map(lambda x: x**2, numbers)

What is the difference between map(), reduce(), and filter() functions in Python?

map(): Applies a function to each item of an iterable and returns an iterator of the results.    
filter(): Constructs an iterator from those elements of an iterable for which a function returns true.    
reduce(): Applies a function of two arguments cumulatively to the items of a sequence, from left to right, so as to reduce the sequence to a single value.  (Note: reduce() is in the functools module in Python 3).

```python
"""Write a Python function that takes a list of numbers as input and returns the
sum of all even numbers in the list."""

def sum_even_numbers(numbers):
    """Calculates the sum of even numbers in a list."""
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum


numbers = [1, 2, 3, 4, 5, 6]
even_sum = sum_even_numbers(numbers)
print(f"The sum of even numbers is: {even_sum}")  
```

```python
"""Create a Python function that accepts a string
and returns the reverse of that string.  """


def reverse_string(input_string):
    """Reverses a given string."""
    return input_string[::-1]


string = "hello"
reversed_string = reverse_string(string)
print(f"The reversed string is: {reversed_string}")  
```

```python
""" Implement a Python function that takes a list of integers and
returns a new list containing the squares of each number."""


def square_numbers(numbers):
    """Squares each number in a list."""
    squared_list = [num**2 for num in numbers]
    return squared_list


numbers = [1, 2, 3, 4]
squared_numbers = square_numbers(numbers)
print(f"The list of squares is: {squared_numbers}")  # Output: [1, 4, 9, 16]
```

```python
"""Write a Python function that checks if a given number is prime or not from 1 to 200.    

 """


def is_prime(number):
    """Checks if a number is prime."""
    if number <= 1:
        return False
    if number <= 3:
        return True
    if number % 2 == 0 or number % 3 == 0:
        return False
    i = 5
    while i * i <= number:
        if number % i == 0 or number % (i + 2) == 0:
            return False
        i += 6
    return True


for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is a prime number")
```

```python
"""Create an iterator class in Python that generates the
Fibonacci sequence up to a specified number of terms. """


class FibonacciIterator:
    """Generates the Fibonacci sequence."""

    def __init__(self, n):
        self.n = n
        self.a = 0
        self.b = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.n:
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return result
        else:
            raise StopIteration


fib_seq = FibonacciIterator(10)  
for num in fib_seq:
    print(num)
```

```python
"""Write a generator function in Python that yields the powers of 2 up 
to a given exponent.    

 """

def powers_of_two(exponent):
    """Generates powers of 2 up to a given exponent."""
    for i in range(exponent + 1):
        yield 2**i


for power in powers_of_two(5):
    print(power)
```

```python
"""Implement a generator function that reads a file line by line and
yields each line as a string."""


def read_file_lines(filename):
    """Reads a file line by line and yields each line."""
    try:
        with open(filename, 'r') as file:
            for line in file:
                yield line.strip()  # Remove leading/trailing whitespace
    except FileNotFoundError:
        yield "File not found."

# Example Usage
# Create a dummy file for testing
with open("my_file.txt", "w") as f:
    f.write("This is line 1\n")
    f.write("This is line 2\n")

for line in read_file_lines("my_file.txt"):
    print(line)

for line in read_file_lines("nonexistent_file.txt"):
    print(line)
```

```python
"""Use a lambda function in Python to sort a list 
of tuples based on the second element of each tuple."""

tuples = [('a', 5), ('b', 2), ('c', 7), ('d', 1)]
sorted_tuples = sorted(tuples, key=lambda x: x[1])  
print(sorted_tuples)  
```

```python
"""Write a Python program that uses map() to convert
a list of temperatures from Celsius to Fahrenheit."""

celsius_temps = [25, 30, 15, 20]
fahrenheit_temps = list(map(lambda c: (9/5) * c + 32, celsius_temps))
print(fahrenheit_temps)  
```

```python
"""Create a Python program that uses filter() to remove all the vowels from a given string.    
"""

def remove_vowels(input_string):
    """Removes vowels from a string."""
    vowels = "AEIOUaeiou"
    return ''.join(filter(lambda char: char not in vowels, input_string))

string = "Hello, World!"
no_vowels = remove_vowels(string)
print(no_vowels)  
```
